```python
col_names =  ['url', 'html', 'label']
Phish_data1  = pd.DataFrame(columns = col_names)
#feature_file_df['RESULT'] = RESULT_df['RESULT'].values
Phish_data1['url']=url_data.values
Phish_data1['html']=html_data.values
Phish_data1['label']='0'
```

```python
title=[]
defined_encoding = []
current_encoding =[]
visible_text =[]
forms = []
images =[]
images_in_forms = []
javascript =[]
logged_links =[]
href_links =[]
iframes =[]
for i in Phish_data1.html:
    page = Webpage(i)
    title.append(page.title)
    defined_encoding.append(page.defined_encoding)
    current_encoding.append(page.current_encoding)
    visible_text.append(page.visible_text)
    forms.append(len(page.forms))
    images.append(len(page.images))
    images_in_forms.append(len(page.images_in_forms))
    javascript.append(len(page.javascript))
    logged_links.append(len(page.logged_links))
    href_links.append(len(page.href_links))
    iframes.append(len(page.iframes))
Phish_data["title"] = title
Phish_data["defined_encoding"] = defined_encoding
Phish_data["current_encoding"] = current_encoding
Phish_data["visible_text"] = visible_text
Phish_data["forms"] = forms
Phish_data["images"] = images
Phish_data["images_in_forms"] = images_in_forms
Phish_data["javascript"] = javascript
Phish_data["logged_links"] = logged_links
Phish_data["href_links"] = href_links
Phish_data["iframes"] = iframes
```
