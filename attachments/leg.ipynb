```python

data_leg = pd.read_csv("leg.csv")
```

```python
data_leg = data_leg[:500]
```

```python

html_data_leg = data_leg.iloc[:,3]
url_data_leg= data_leg.iloc[:,1]
```

```python
col_names =  ['url', 'html', 'label']
Leg_data1  = pd.DataFrame(columns = col_names)

Leg_data1['url']=url_data_leg.values
Leg_data1['html']=html_data_leg.values
Leg_data1['label']='1'
```

```python
Leg_data1
```

```python
Leg_data = pd.DataFrame()
```

```python

forms = []
javascript =[]
logged_links =[]
href_links =[]
iframes =[]
for i in Leg_data1.html:
    page = Webpage(i)
    forms.append(len(page.forms))
    javascript.append(len(page.javascript))
    logged_links.append(len(page.logged_links))
    href_links.append(len(page.href_links))
    iframes.append(len(page.iframes))

Leg_data["forms"] = forms
Leg_data["javascript"] = javascript
Leg_data["logged_links"] = logged_links
Leg_data["href_links"] = href_links
Leg_data["iframes"] = iframes
```

```python
j = 0
ip=[]
ip2 = []
www=[]
www2 =[]

for i in Leg_data1.url:
    url_f = Url(i)
    
    ip.append(url_f.ip)
    ip2.append(url_f.ip)
    if None in  ip:
        ip[j] = 0
    else:
        ip[j] = 1
    #print(ip)
    www.append(url_f.www)
    www2.append(url_f.www)
    #print(www)
    if True in  www:
        www[j] = 'y'
    else:
        www[j] = 'n'
    j= j+1
#print(www)
for i in range(j):
    #print(i)
    if www[i]=='y':
        www[i] = 1
    else:
        www[i] = 0
#print(www)
    

Leg_data["ip"] = ip
#Leg_data["ip2"] = ip2
Leg_data["www"] = www
#Leg_data["www2"] = www2

```

```python

protocol = []
nof_vowels = []
nof_consonants = []
nof_special_sym = []
nof_digits = []
nof_alpha = []
nof_alphanumeric = []
nof_ques = []
nof_dollar = []
nof_percent = []
nof_at = []
nof_dash = []
nof_dots = []
nof_dir = []
nof_www = []
nof_http = []
nof_eq = []
dash_in_path = []
terms_in_url = []



for i in Leg_data1.url:
    protocol.append(fe.protocol(i))
    nof_vowels.append(fe.nof_vowels(i))
    nof_consonants.append(fe.nof_consonants(i))
    nof_special_sym.append(fe.nof_special_sym(i))
    nof_digits.append(fe.nof_digits(i))
    nof_alpha.append(fe.nof_alpha(i))
    nof_alphanumeric.append(fe.nof_alphanumeric(i))
    nof_ques.append(fe.nof_ques(i))
    nof_dollar.append(fe.nof_dollar(i))
    nof_percent.append(fe.nof_percent(i))
    nof_at.append(fe.nof_at(i))
    nof_dash.append(fe.nof_dash(i))
    nof_dots.append(fe.nof_dots(i))
    nof_dir.append(fe.nof_dir(i))
    nof_www.append(fe.nof_www(i))
    nof_http.append(fe.nof_http(i))
    nof_eq.append(fe.nof_eq(i))
    dash_in_path.append(fe.dash_in_path(i))
    terms_in_url.append(fe.terms_in_url(i))
    
Leg_data["protocol"] = protocol
Leg_data["nof_vowels"] = nof_vowels
Leg_data["nof_consonants"] = nof_consonants
Leg_data["nof_special_sym"] = nof_special_sym
Leg_data["nof_digits"] = nof_digits
Leg_data["nof_alpha"] = nof_alpha
Leg_data["nof_alphanumeric"] = nof_alphanumeric
Leg_data["nof_ques"] = nof_ques
Leg_data["nof_dollar"] = nof_dollar
Leg_data["nof_percent"] = nof_percent
Leg_data["nof_at"] = nof_at
Leg_data["nof_dash"] = nof_dash
Leg_data["nof_dots"] = nof_dots
Leg_data["nof_dir"] = nof_dir
Leg_data["nof_www"] = nof_www
Leg_data["nof_http"] = nof_http
Leg_data["nof_eq"] = nof_eq
Leg_data["dash_in_path"] = dash_in_path
Leg_data["terms_in_url"] = terms_in_url
```
